```python
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
input_size = 32
patch_size = 8

loss = torch.randn((input_size, input_size))
loss = loss * loss
```

```python
loss_per_patch = F.conv2d(loss[None, None], torch.ones(1,1, patch_size, patch_size))
loss_per_patch = loss_per_patch[0,0] / loss_per_patch.sum()
```

```python
def sample_index(p):
    #https://stackoverflow.com/questions/61047932/numpy-sampling-from-a-2d-numpy-array-of-probabilities
    i = np.random.choice(np.arange(p.size), p=p.ravel())
    return np.unravel_index(i, p.shape)
```

```python
#show loss and loss_per_patch next to each other
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(loss)
plt.colorbar()
plt.title('loss')
plt.subplot(1, 2, 2)
plt.imshow(loss_per_patch)
plt.title('loss_per_patch')
plt.colorbar()
plt.show()
```

```python
sample_index(loss_per_patch.numpy())
```
