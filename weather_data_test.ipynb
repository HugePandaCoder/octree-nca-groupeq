```python
import numpy as np
import xarray as xr
import matplotlib
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.feature import BORDERS
import numpy as np

data = xr.open_dataset("/local/scratch/clmn1/data/CloudCast/TestCloud.nc")
data = data.where(data>4)
data = data-4
data = data.fillna(0)
```
